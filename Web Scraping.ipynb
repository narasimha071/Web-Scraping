**Extracting the total data from website**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL with a placeholder for page number
BASE_URL = 'https://www.scrapethissite.com/pages/forms/?page_num={}'

# Initialize an empty list to store rows
all_rows = []

# Loop through all 24 pages
for page in range(1, 25):  # Pages 1 to 24
    print(f"Scraping page {page}...")

    # Fetch the content of the page
    response = requests.get(BASE_URL.format(page))
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table
    table = soup.find('table', class_='table')

    # Extract rows from the table
    for row in table.find_all('tr')[1:]:  # Skip the header row
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]  # Clean text
        all_rows.append(cols)  # Append row data to all_rows list

# Extract column headers (only from the first page)
headers = [header.text.strip() for header in table.find_all('th')]

# Create a DataFrame
df = pd.DataFrame(all_rows, columns=headers)

# Display the DataFrame
print(df)

# Save to CSV
df.to_csv('scraped_data.csv', index=False)
print("Data saved to 'scraped_data.csv'")
```

**3.Count the win%,Wins,Losses by Each Team**

```python
import pandas as pd
numeric_columns=['Wins','Losses','Win %',]
for i in numeric_columns:
    New_df.loc[:,i] = pd.to_numeric(New_df[i],errors='coerce')
team= New_df.groupby('Team Name').agg(
    Total_Wins=('Wins','sum'),
    Total_Losses=('Losses','sum'),
    Average_Win_percent=('Win %','mean')
).reset_index()

team_New=team.sort_values(by='Team Name',ascending=True)
print(team_New)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.bar(team_New['Team Name'],team_New['Total_Wins'],label='Wins',color='green')
plt.bar(team_New['Team Name'],team_New['Total_Losses'],label='Losses',color='red',alpha=0.7)

plt.title('Wins and Losses by Each Team',fontsize=14)
plt.xlabel('Team Names',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=45,ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
